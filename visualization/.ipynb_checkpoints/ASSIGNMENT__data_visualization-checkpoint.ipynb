```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Read the CSV file
data = pd.read_csv('airquality.csv')
```

```python
data
```

# PIE CHART

```python
labels = ['Ozone', 'Solar Radiation', 'Wind', 'Temperature']
sizes = [data['Ozone'].mean(), data['Solar.R'].mean(), data['Wind'].mean(), data['Temp'].mean()]
colors = ['red', 'gold', 'purple', 'brown']
explode = (0.1, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Average data')
plt.show()
```

# BAR PLOT

```python
h = data.iloc[0:20, 4]
y_pos = np.arange(len(h))
v = range(1, 21)
plt.bar(y_pos, h, align='center', alpha=0.5)
plt.xticks(y_pos, v)
plt.ylabel('Temperature')
plt.xlabel("Days")
plt.title('Temperature for 1st-20th May')
plt.show()
```

# LINE GRAPH

```python
h = data.iloc[0:15, 1]
v = data.iloc[0:15, 3]
plt.plot(h, label='ozone', marker='o', linestyle='dotted')
plt.plot(v, label='wind', marker='o', linestyle='dashed')
plt.ylim(0, 60)
plt.legend()
plt.title("Ozone and Wind chart for 1st-15th May")
plt.ylabel("Value Range")
plt.xlabel("Days")
plt.show()
```

```python
sns.set(style = "ticks")
sns.pairplot(data)
plt.show()
```

# HEAT MAP

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights)
```

```python
Theory Visualization:
It may sometimes seem easier to go through a set of data points and build insights from it but usually this 
process may not yield good results. There could be a lot of things left undiscovered as a result of this process. 
Additionally, most of the data sets used in real life are too big to do any analysis manually.
Data visualization is an easier way of presenting the data, however complex it is, to analyze trends and 
relationships amongst variables with the help of pictorial representation.
The following are the advantages of Data Visualization
ï‚· Easier representation of compels data
ï‚· Highlights good and bad performing areas
ï‚· Explores relationship between data points
ï‚· Identifies data patterns even for larger data pointsVisualization should have:
ï‚· Appropriate usage of shapes, colors, and size while building visualization
ï‚· Plots/graphs using a co-ordinate system are more pronounced
ï‚· Knowledge of suitable plot with respect to the data types brings more clarity to the information
ï‚· Usage of labels, titles, legends and pointers passes seamless information the wider audience
Visualization libraries in python:
There are a lot of python libraries which could be used to build visualization like matplotlib, vispy, bokeh, 
seaborn, pygal, folium, plotly, cufflinks, and networkx. Of the many, matplotlib and seaborn seems to be very 
widely used for basic to intermediate level of visualizations.

1.Matplotlib
It is an library in Python for 2D plots of arrays, It is a multi-platform data visualization library built on
NumPy arrays and designed to work with the broader SciPy stack.
It is well maintained visualization output with high quality graphics draws a lot of users to it. Basic as well as 
advanced charts could be very easily built from the users/developers point of view, since it has a large 
community support, resolving issues and debugging becomes much easier

2.Seaborn
This library sits on top of matplotlib. Means it has some flavors of matplotlib while from the visualization 
point, it is much better than matplotlib and has added features as well.
Benefits:
ï‚· Built-in themes aid better visualization
ï‚· Statistical functions aiding better data insights
ï‚· Better aesthetics and built-in plots
ï‚· Helpful documentation with effective examples
```
